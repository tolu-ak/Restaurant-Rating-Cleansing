```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('california_restaurants_cleaned.csv')
df.head()
```

```python
plt.figure(figsize=(8, 5))
df['Rating'].value_counts().sort_index().plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('rating_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
price_counts = df['Price Range'].value_counts()
price_counts.index = [
    'Inexpensive' if x == '$'
    else 'Moderate' if x == '$$'
    else 'Expensive' if x == '$$$'
    else 'Very Expensive' for x in price_counts.index
]
price_counts.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Distribution of Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('price_range_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['Number of Reviews'].dropna().plot(kind='hist', bins=30, color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('reviews_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Category 1'].value_counts().head(15).plot(kind='bar', color='mediumpurple', edgecolor='black')
plt.title('Top 15 Food Categories')
plt.xlabel('Category')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('category_distribution.png')
plt.show()
```
